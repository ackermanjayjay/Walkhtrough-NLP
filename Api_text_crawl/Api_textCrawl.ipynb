> Library `request`

# import library

```python
import requests as req
import pandas as pd
```

```python
# Tutorial https://pypi.org/project/requests/
crawl_text=req.get("https://dummyjson.com/products")
```

```python
crawl_text.status_code
```

Status code 200 Everything went okay, and the result has been returned 

```python
data_crawl=crawl_text.json() # Optional arguments that json.loads takes.
```

> API best practise use type data Json

>but in library request can use .text 

>but in practice just use .json 

```python
data_crawl
```

```python
data_crawl_to_dataFrame=data_crawl['products']
```

```python
data_crawl_fix=pd.DataFrame(data_crawl_to_dataFrame)
```

```python
data_crawl_fix.to_csv("crawl_dummy_json.csv",index=False)
```
