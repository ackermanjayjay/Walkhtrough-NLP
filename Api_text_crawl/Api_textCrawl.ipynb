## Casefolding

```python
import re
```

```python
def caseFolding(text):
  text=text.lower()
  text=re.sub(r"[^a-zA-Z]"," ",text) # Menghapus tanda baca dan angka
  text=text.strip()
  return text
```

```python
raw=dataDipakai["description"][1]
print("non_casefolding :",raw)
caseFoldingText=dataDipakai["description"][1]
print(f'case folding ->{caseFolding(caseFoldingText)}')
```

## Tokenization

```python
def tokenization(text):
  text=word_tokenize(text)
  return text
```

```python
tokenization(raw)
```

## Stopwords

```python
stopwords_En=stopwords.words('english')
stp=stopwords_En
```

```python
def remove_stop_words(text):
  clean_words = []
  text=text.split()
  for word in text:
      if word not in stp:
          clean_words.append(word)
  # perulangan_text=[word for word in text]
  # if perulangan_text not in stp: 
  #   clean_words.append(perulangan_text)
  return " ".join(clean_words)
```

```python
raw=data["description"].iloc[5]
stopWordsText=caseFolding(raw)
```

```python
stopWordsText
```

```python
remove_stop_words(stopWordsText)
rm_stopWords=remove_stop_words(stopWordsText)
```

## Stemming

```python
from nltk.stem.porter import *
```

```python
stemmer = PorterStemmer()
```

```python
def stemmingEn(text):
  text=stemmer.stem(text)
  return text
```

## Word normalization

```python
# Buat korpus bank 
korpus_word_normal={
    "kata":["appl"],
    "perbaikan":["apple"]
}
key_norm=pd.DataFrame(korpus_word_normal)
```

```python
def text_normalize(text):
  text = ' '.join([key_norm[key_norm['kata'] == word]['perbaikan'].values[0] if (key_norm['kata'] == word).any() else word for word in text.split()])
  text = str.lower(text)
  return text
```

## Text preprocessing pipeline

```python
def text_preprocessing_process(text):
  text = caseFolding(text)
  text = remove_stop_words(text)
  text = stemmingEn(text)
  text=text_normalize(text)
  return text
```

```python
%%time 
dataDipakai["clean text"]=data["description"].apply(text_preprocessing_process)
```

```python
text_algorithm_naives_bayes=MultinomialNB()
```

```python
model = text_algorithm_naives_bayes.fit(x_train, y_train)
```

```python
#Prediksi
predicted = model.predict(x_test)

#Hasil ini akan kita bandingkan dengan nilai y_test (labeling dari manusia)
predicted
```

```python
#Library evaluasi
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#Compute performance manually
NewprediksiBenar = (predicted == y_test).sum()
NewprediksiSalah = (predicted != y_test).sum()
    
print("prediksi benar: ", NewprediksiBenar, " data")
print("prediksi salah: ", NewprediksiSalah, " data")
print("Akurasi Algoritme: ", NewprediksiBenar/(NewprediksiBenar+NewprediksiSalah)*100,"%")

CM = confusion_matrix(y_test,predicted)

TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]
precision    = TP/(TP+FP)
recall       = TP/(TP+FN)
print ("TRUE NEGATIVE (TN):",TN)
print ("FALSE NEGATIVE (FN):",FN)
print ("TRUE POSITIVE (TP):",TP)
print ("FALSE POSITIVE (FP):",FP)
print ("PRECISION:",precision*100,"%")
print ("RECALL:",recall*100,"%")
print(classification_report(y_test,predicted))
```
